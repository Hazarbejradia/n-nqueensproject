```python
import copy
import random
def saisie(): #fonction pour la saisie de n
 
    while True:
        try:
            n = int(input('Donner le nombre de Reines ? n = '))
            if n<4 and n>1 : #nombre de reine =3 ou <3 ( pas de solution pour n=3 et n=2 et pour n=1 problème simple (une case pour n=1)
                print("donner une valeur superieur ou égal  4")
                continue
            return n 
        except ValueError:
              print("valeur n'est pas valide !! ")
n = saisie()

def Remplir(n): #fonction pour remplissage de n*n à l'aide d'un tableau dynamique
    
    T = [" "]*n
    for i in range(n):
        T[i] = [" "]*n
    return T

def Est_Libre(T, row, col, n): #fonction pour verifier la disponibilité de case libre dans la matrice 

    for j in range(col): 
        if T[row][j] == "R":
            return False #parcourir tous les colonnes de Tableau et verifier si on a une case occupé par R (colonne interdite) renvoie false
    
    i, j = row, col
    while i >= 0 and j >= 0: #parcourir si sur le diagonale on a une case occupée par R retourner false
        if T[i][j] == "R": 
            return False
        i=i-1 #compteur -1 ( ligne -1 )
        j=j-1 # colonne -1
    
    x, y = row,col
    while x < n and y >= 0: 
        if T[x][y] == "R":
            return False 
        x=x+1 #ligne +1
        y=y-1
    
    return True #case libre

def ajouter_solution(T):

    global solutions
    saved_board = copy.deepcopy(T)
    solutions.append(saved_board)


    
T = Remplir(n)

solutions = []

def placer_reine(T, col, n):
    
    if col >= n:
        return
    
    for i in range(n):
        if Est_Libre(T, i, col, n):
            T[i][col] = "R"
            if col == n-1:
                ajouter_solution(T)
                T[i][col] = " "
                
            placer_reine(T, col+1, n) 
            T[i][col] = " "
def Afficher_solution(solutions, n):
 
    x = random.randint(0,len(solutions)-1) # ajouter une solution aléatoire selon nombre des solutions

    for row in solutions[x]:
        
        print("| " + " | " .join(row) +" | ") #pour l'affichage et emplacement des reines      
placer_reine(T, 0, n)
print()
Afficher_solution(solutions, n)
print()
p=len(solutions)
print("Nombre de solutions totale est ègal  "  + (str)(p)+ "")
```
